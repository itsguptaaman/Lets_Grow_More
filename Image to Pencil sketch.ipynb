```python
#Importing Libraries
import cv2
from PIL import Image
from IPython.display import display
```

```python
#  reading the image
image = cv2.imread(r'C:\Users\Aman\Downloads\2022-02-04.jpg')
```

#### **Displaying the image**

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(image))
```

#### **Converting into grey scale image**

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_image))
```

```python
inv_grey_image = 255 - grey_image
display(Image.fromarray(inv_grey_image))
```

#### Bluring the image

```python
blur_image = cv2.GaussianBlur(inv_grey_image, (35, 35), 0)
```

```python
display(Image.fromarray(blur_image))
```

```python
inverted_blur_image = 255 - blur_image
```

```python
display(Image.fromarray(inverted_blur_image))
```

#### Pencil sketch image

```python
pencil_sketch = cv2.divide(grey_image, inverted_blur_image, scale=256.0)
```

```python
cv2.imshow("Sketch Image", pencil_sketch)
display(Image.fromarray(pencil_sketch))
```

```python
display(Image.fromarray(image))
display(Image.fromarray(pencil_sketch))
```
